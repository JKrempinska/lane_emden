## Newtonowski model gwiazdy
Newtonowskim modelem gwiazdy określamy samograwitujący, sferycznie symetryczny, politropowy płyn. Do dodatkowych założeń należy bezrotacyjność, niemagnetyczność, zerowa wartość siły wypadkowej oraz przyspieszenia wypadkowego. [2]

Aby zbadać równowagę takiego ciała zaczniemy od zbadania ciągłości masy. 
Masa jednostkowa zgromadzona w powłoce o promieniu $r$ oraz nieskończenie małej grubości $dr$ określana jest wzorem 

$$
{\displaystyle m_{r+dr}-m_r=dm_r=4\pi r^{2}\rho(r) dr}
$$

Następnie przejdziemy do określenia równowagi hydrostatycznej. 
Rozważmy jednostkowe pole na powierzchni powłoki o promieniu $r$. Określmy również jednostkową opisywaną przez to pole oraz nieskończenie mały odcinek $dr$. [2] W kierunku wewnętrznym na tą objętość działa siła grawitacyjna określona wzorem

$$
{\displaystyle \rho g dr = \rho \frac{Gm_r}{r^{2}} dr}
$$

W przeciwnym kierunku działa ciśnienie $P$, przeciwdziałające zapadaniu grawitacyjnemu, które musi mieć większą wartość niż działająca siła grawitacyjna. [2] Zatem 

$$
{\displaystyle P(r)-P(r+dr)=-\frac{dP}{dr} dr}
$$

$$
{\displaystyle \rho \ddot{r}=-\frac{dP}{dr}- \rho \frac{Gm_r}{r^{2}} }
$$

Zgodnie z założeniami, siły wypadkowe są równe zero, więc $\ddot{r}=0$. [2] Otrzymujemy zatem równanie równowagi hydrostatycznej:

$$
{ \displaystyle \frac {dP}{dr}=-{\frac {Gm_r}{r^{2}} \rho=-g \rho }}
$$

### Równanie Lane - Emdena
Omawiane wyżej równania gwiazdy newtonowskiej:
$$
{ \displaystyle \frac {dP}{dr}=-{\frac {Gm_r}{r^{2}} \rho }}
$$

$$
{ \displaystyle {\frac{dm}{dr}=4\pi r^{2}\rho}}
$$

można przekształcić do jednego rówania

$$ {\displaystyle {\frac {1}{r^{2}}}{\frac {d}{dr}}({\frac {r^{2}}{\rho }}{\frac {dP}{dr}})=-4\pi G\rho (r)} 
$$

W równaniach mamy trzy niewiadome: ciśnienie $P$, gęstość $\rho$ oraz masę w funkcji promienia $r$, dlatego aby rozwiązać zagadninnie należy określić kolejne założenie lub równanie stanu. [2]
Zakładamy więc politropię gazu gwiazdy, czyli stały związek pomiędzy ciśnieniem oraz objętością płynu. Równanie stanu politropy wygląda następująco 

$$
{\displaystyle {P={K\rho(r) ^{1+{\frac {1}{n}}}}}}
$$

gdzie $n$ jest wykładnikiem politropy, a $K$ jest stałe. 
Wykorzystując powyższe równanie, całe zagadnienie możemy sprowadzić do pojedyńczego równania, co zostanie przedstawione w dalszej części. Równanie zostało nazwane nazwiskami astrofizyków, którzy je opracowali - Jonathana Homera Lane'a oraz Roberta Emdena [2]


## Wyprowadzenie równania Lane-Emdena [3]
Zakładamy, że mamy gwiazdę w równowadze hydrodynamicznej, o stosunkowo małej masie, sferycznie symetryczną. Niech $r$ - promień gwiazdy, $m$ - masa, $\rho$ - gęstość, $P$ - ciśnienie gazu, $n$ - indeks adiabatyczny. Wiemy, że rozkład masy gwiazdy możemy opisać równaniem

$$
\begin{equation} \label{eq:1}
{\frac {dm}{dr}}=4\pi \rho (r)r^{2}
\end{equation}
$$

Będziemy korzystać również z równania opisującego równowagę hydrostatyczną 

$$
\begin{equation} \label{eq:2}
{{\frac {1}{\rho }}{\frac {dP}{dr}}=-{\frac {Gm}{r^{2}}}}
\end{equation}
$$

oraz równania stanu 

$$
\begin{equation} \label{eq:3}
{P={K\rho(r) ^{1+{\frac {1}{n}}}}}
\end{equation}
$$

Aby wyprowadzić wzór zaczniemy od zróżniczkowania obustronnie równania \eqref{eq:2};

$$
{\displaystyle {\frac {d}{dr}}\left({\frac {1}{\rho }}{\frac {dP}{dr}}\right)={\frac {2Gm}{r^{3}}}-{\frac {G}{r^{2}}}{\frac {dm}{dr}}}
$$

Zauważmy, że po prawej stronie możemy podstawić rownania \eqref{eq:1} i \eqref{eq:2} aby otrzymać

$$
{\displaystyle {\frac {d}{dr}}\left({\frac {1}{\rho }}{\frac {dP}{dr}}\right)=-{\frac {2}{\rho r}}{\frac {dP}{dr}}-4\pi G\rho}
$$

Teraz przenosimy wyrażenie $-{\frac {2}{\rho r}}{\frac {dP}{dr}}$ na drugą stronę, a natępnie mnożymy całość przez $r^{2}$ i uzyskujemy

$$
{\displaystyle {\frac {d}{dr}}\left({\frac {r^{2}}{\rho }}{\frac {dP}{dr}}\right)=-4\pi Gr^{2}\rho }
$$

W kolejnym kroku dzielimy obustronnie przez $r^{2}$,a następnie korzystamy z równania stanu \eqref{eq:3} oraz równania $\rho=\rho_c\theta^{n}$, gdzie $\rho_c$ to gęstość centralna, a $\theta$ gęstość znormalizowana. Otrzymujemy

$$
{\displaystyle {\frac {1}{r^{2}}}{\frac {d}{dr}}\left(r^{2}K\rho _{c}^{\frac {1}{n}}(n+1){\frac {d\theta }{dr}}\right)=-4\pi G\rho _{c}\theta ^{n}}
$$

Następnie stosujemy podstawienie $r=\alpha\xi$, gdzie 

$$
{\displaystyle {\alpha ^{2}=(n+1)K\rho _{c}^{{\frac {1}{n}}-1}/4\pi G}}
$$

Ostatecznie otrzymujemy końcową postać równania Lane-Emdena

$$
{\displaystyle {\frac {1}{\xi ^{2}}}{\frac {d}{d\xi }}\left({\xi ^{2}{\frac {d\theta }{d\xi }}}\right)+\theta ^{n}=0}
$$

Warunkami brzegowymi tego równania są

$$
\displaystyle{\theta(0)=1}
$$
$$
\displaystyle{\theta'(0)=0}
$$

Możemy zauważyć, że równanie Lane'a - Emdenego jest bezwymiarowe. Określone w nim $\xi$ jest bezwymairowym promieniem, natomiast $\theta$ jest w relacji z gęstością, a dalej z ciśnieniem. Pojawiające się w równaniu $n$ jest wykładnikiem politropy, charakterystycznym dla rozważanego gazu. 


## Rozwiązania analityczne
Z fizycznego punktu widzenia [4] istotne są wartości $n$ należące do przedziału $[0,5]$. Jednak rozwiązania analityczne istnieją tylko dla $n=0,1$ oraz $5$. 
### Rozwiązanie dla $n=0$
W tym przypadku równanie Lane-Emdena ma postać

$$
{\displaystyle {\frac {1}{\xi ^{2}}}{\frac {d}{d\xi }}\left(\xi ^{2}{\frac {d\theta }{d\xi }}\right)+1=0}
$$

Przenosimy $1$ na drugą strone, a następnie mnożymy obustronnie przez $\xi^{2}$ i dalej

$$
\int d\left( \xi ^{2}{\frac {d\theta }{d\xi }}\right) = -\int \xi^{2} \,d\xi
$$

Po całkowaniu i uporządkowaniu otrzymujemy równanie 

$$
{\frac {d\theta}{d\xi}} = {\frac {c_1-{{\frac{1}{3}}\xi^3}}{\xi^{2}}}
$$

Ponownie całkujemy i ostateczne mamy

$$
\theta (\xi)=\theta_{0}-{\frac {c_{1}}{\xi }}-{\frac {1}{6}}\xi ^{2}
$$

Z warunków brzegowych wyznaczamy stałą $c_{1} = 1$ i ostateczne rozwiązanie dla $n=0$ to

$$
\theta_{1}(\xi)=1-{\frac {1}{6}}\xi ^{2}
$$



```python
import numpy as np
ksi = [1e-50]
for i in np.arange(0.1, 14, 0.25):
    ksi.append(i)

#Wykres rozwiązania dla n=0
theta = []
for i in ksi:
    theta.append(1 - i**2 / 6)
plt.plot(ksi, theta)
plt.title('Rozwiązanie równania Lane-Emdena dla n=0')
plt.xlabel('ξ')
plt.ylabel('θ(ξ)')
plt.ylim(top=1,bottom=-0.6)
plt.grid(True)
plt.show()
```

### Rozwiązanie dla $n=1$
Teraz równanie przyjmuje postać

$$
{\displaystyle {\frac {1}{\xi ^{2}}}{\frac {d}{d\xi }}\left(\xi ^{2}{\frac {d\theta }{d\xi }}\right)+\theta=0}
$$

Po obustronnym pomnożeniu przez $\xi^{2}$ mamy

$$
{\displaystyle {\frac {d}{d\xi }}\left(\xi ^{2}{\frac {d\theta }{d\xi }}\right)+\theta\xi^{2}=0}
$$

To równanie można zapisać [5] w postaci

$$
{\displaystyle {\frac {d}{dr}}\left(r ^{2}{\frac {dR}{dr}}\right)+R\left[k^{2}r^{2}-m(m+1)\right]=0}
$$

Widać teraz, że jest to równanie Bessela z $k=1$ i $m=0$. Stąd rozwiązanie będzie miało postać

$$
{\displaystyle \theta (\xi )=A{\frac {\sin \xi }{\xi }}+B{\frac {\cos \xi }{\xi }}}
$$

Z warunku $\theta(0)=1$ ostatecznie wyznaczamy rozwiązanie dla $n=1$

$$
\theta_{2} (\xi )={\frac {\sin \xi }{\xi }}
$$



```python
#Wykres rozwiązania dla n=1
theta = []
for i in ksi:
    theta.append((np.sin(i))/i)
plt.plot(ksi, theta)
plt.title('Rozwiązanie równania Lane-Emdena dla n=1')
plt.xlabel('ξ')
plt.ylabel('θ(ξ)')
plt.grid(True)
plt.ylim(top=1,bottom=-0.6)
plt.show()
```

### Rozwiązanie dla $n=5$
Równanie Lane-Emdena przyjmuje teraz postać

$$
{\displaystyle {\frac {1}{\xi ^{2}}}{\frac {d}{d\xi }}\left(\xi ^{2}{\frac {d\theta }{d\xi }}\right)+\theta ^{5}=0}
$$

Aby je rozwiązać stosujemy podstawienie [6]

$$
\theta=\frac{z}{\sqrt{2\xi}}
$$

$$
t=-\ln{\xi}
$$

Otrzymujemy postać równania

$$
\frac{d^{2}z}{dt^{2}}=\frac{1}{4}z\left(1-z^{4}\right)
$$

Po scałkowaniu mamy

$$
\left(\frac{dz}{dt}\right)^{2}=\frac{1}{12}\left(-z^{6}+3x^{2}+C\right)
$$

Po dalszych przekształceniach [5] otrzymujemy ostateczne rozwiązanie dla $n=5$ postaci

$$
{\displaystyle \theta_3 (\xi )={\frac {1}{\sqrt {1+\xi ^{2}/3}}}}
$$


```python
#Wykres rozwiązania dla n=5
theta = []
for i in ksi:
    theta.append(1/(np.sqrt(1+((i)**2)/3)))
plt.plot(ksi, theta)
plt.title('Rozwiązanie równania Lane-Emdena dla n=5')
plt.xlabel('ξ')
plt.ylabel('θ(ξ)')
plt.grid(True)
plt.ylim(top=1,bottom=-0.6)
plt.show()
```

## Rozwiązania numeryczne

### Metoda Eulera
Dla pozostałych wartości $n$ należy zastosować metody numeryczne. Pierwszą jaką przedstawimy jest metoda Eulera. Przybliżymy rozwiązania równania Lane-Emdena poprzez założenie, że prędkość zmiany funkcji w danym punkcie jest równa przyrostowi funkcji w tym punkcie, co opisują wzory

$$
\theta_{i+1}=\theta_{i}+h \frac{d\theta}{d\xi}_{i}
$$

$$
\frac{d\theta}{d\xi}_{i+1}=\frac{d\theta}{d\xi}_{i}+h\left(-\frac{1}{\xi_{i}^{2}}\frac{d}{d\xi}\left(\xi_{i}^{2}\frac{\theta}{\xi}\right)_{i}-\theta_{i}^{n}\right)
$$


```python
import matplotlib.pyplot as plt

def lane_emden_euler(n, xi_max, h):

    xi_values = []
    theta_values = []

    xi = 1e-10
    theta = 1.0
    dtheta_dxi = 0.0
    while xi <= xi_max:
        xi_values.append(xi)
        theta_values.append(theta)

        theta_next = theta + h * dtheta_dxi
        if xi != 0:
            dtheta_dxi_next = dtheta_dxi + h * (-1 / xi**2 * (2 * xi * dtheta_dxi + xi**2 * theta**n))
        else:
            dtheta_dxi_next = 0

        xi += h
        theta = theta_next
        dtheta_dxi = dtheta_dxi_next
    return (xi_values, theta_values)

plt.plot(lane_emden_euler(0, 14, 0.1)[0],lane_emden_euler(0, 14, 0.1)[1],label='n=0')
plt.plot(lane_emden_euler(1, 14, 0.1)[0],lane_emden_euler(1, 14, 0.1)[1],label='n=1')
plt.plot(lane_emden_euler(2, 14, 0.1)[0],lane_emden_euler(2, 14, 0.1)[1],label='n=2')
plt.plot(lane_emden_euler(3, 14, 0.1)[0],lane_emden_euler(3, 14, 0.1)[1],label='n=3')
plt.plot(lane_emden_euler(4, 14, 0.1)[0],lane_emden_euler(4, 14, 0.1)[1],label='n=4')
plt.plot(lane_emden_euler(5, 14, 0.1)[0],lane_emden_euler(5, 14, 0.1)[1],label='n=5')

plt.title('Rozwiązanie równania Lane-Emdena dla różnych n')
plt.xlabel('ξ')
plt.ylabel('θ(ξ)')
plt.ylim(top=1,bottom=-0.6)
plt.legend()
plt.grid(True)
plt.show()
```

### Metoda Rungego-Kutty czwartego rzędu

Drugą metodą jest metoda Rungego - Kutty. Polega ona na dokładniejszym przybliżeniu pochodnej funkcji poprzez interpolację pochodnej funkcji w danym punkcie za pomocą pochodnych funkcji w kilku innych punktach. Wzór ogólny jest następujący:

$$
y_{i+1}=y_{i}+h\phi
$$

Metoda Rungego Kutty 4 rzędu polega na obliczeniu kolejno danych wartości funkcji oraz podstawienia ich do wzory ogólnego

$$
\begin{aligned}
k_1 &= f(x_i, \ y_i) \\
k_2 &= f \left(x_i + \frac{h}{2}, \ y_i + \frac{hk_1}{2} \right) \\
k_3 &= f \left(x_i + \frac{h}{2}, \ y_i + \frac{hk_2}{2} \right) \\
k_4 &= f(x_i + h, \ y_i + hk_3) \\
y_{i+1} &= y_i + \frac{1}{6} \, (k_1 + 2k_2 + 2k_3 + k_4)h.
\end{aligned}
$$

Gdzie:
- $y_i$ jest wartością funkcji zależnej w punkcie $x_i$
- $h$ jest krokiem numerycznym
- $f(x_i, \ y_i)$ jest wartością pochodnej funkcji zależnej w punkcie $x_i$

Chcąc rozwiązać równanie różniczkowe drugiego rzędu, jakim jest równanie Lane'a - Emdena, musimy zapisać je jako dwa równania różniczkowe pierwszego rzędu. 
Podstawiając do wzoru 

$$
{\displaystyle {\frac {d}{d\xi }}\left({\xi ^{2}{\frac {d\theta }{d\xi }}}\right)= -{\xi ^{2}}\theta ^{n}}
$$

$ x=\xi, y=\theta, z=\frac{d\theta}{d\xi} $, otrzymujemy 

$$
{\displaystyle {\frac {d}{dx }}\left({x ^{2}{\frac {dy }{dx }}}\right)= -{x ^{2}}y ^{n}}
$$

$$
{\displaystyle {x^{2}}{\frac {d^{2}y}{dx^{2} }}+2x{\frac {dy }{dx }}}+x^{2}y^{n}=0
$$

Wtedy otrzymamy
$$
\begin{aligned}
\frac{dy}{dx} &= z \\
\frac{dz}{dx} &= -\frac{1}{x^{2}} \left(2xz+x^{2}y^{n} \right) \\
\end{aligned}
$$

Możemy zatem przejść do rozwiązania równań. 

```python
def dy_dx(x, y, z):
    return z

def dz_dx(x, y, z, n):
    return -y**n-2/x*z

# integrate
def integrate(x_0, y_0, z_0, n, h):
    k_0 = h * dy_dx(x_0, y_0, z_0)
    l_0 = h * dz_dx(x_0, y_0, z_0, n)
    k_1 = h * dy_dx(x_0+0.5*h, y_0+0.5*k_0, z_0+0.5*l_0)
    l_1 = h * dz_dx(x_0+0.5*h, y_0+0.5*k_0, z_0+0.5*l_0, n)
    k_2 = h * dy_dx(x_0+0.5*h, y_0+0.5*k_1, z_0+0.5*l_1)
    l_2 = h * dz_dx(x_0+0.5*h, y_0+0.5*k_1, z_0+0.5*l_1, n)
    k_3 = h * dy_dx(x_0+h, y_0+k_2, z_0+l_2)
    l_3 = h * dz_dx(x_0+h, y_0+k_2, z_0+l_2, n)
    x_1 = x_0 + h
    y_1 = y_0 + 1/6 * (k_0+2*k_1+2*k_2+k_3)
    z_1 = z_0 + 1/6 * (l_0+2*l_1+2*l_2+l_3)
    return (x_1, y_1, z_1)
    
def rk_integrate(n, h, x_max):
    x_0 = 1e-10 
    y_0 = 1.0
    z_0 = 0

    xs = [x_0]
    ys = [y_0]
    while y_0>0 and x_0<x_max:
        x_0, y_0, z_0 = integrate(x_0, y_0, z_0, n, h)
        if type(y_0) == complex:
            break
        xs.append(x_0)
        ys.append(y_0)
    return (xs, ys)

def rk_plot(n, x_max):
    h=0.1
    xis, thetas = rk_integrate(n, h, x_max)
    plt.plot(xis, thetas, label = "n="+str(n))  
    plt.xlabel(r'$\xi$')
    plt.ylabel(r'$\theta$')
    
def rk_plot_wl(n, x_max):
    h=0.1
    xis, thetas = rk_integrate(n, h, x_max)
    plt.plot(xis, thetas, label = "RK4")  
    plt.xlabel(r'$\xi$')
    plt.ylabel(r'$\theta$')
    
def rk_plot_show(n, x_max):
    for i in range(n):
        rk_plot(i, x_max)
    plt.title('Rozwiązanie równania Lane-Emdena dla różnych n')
    plt.legend()
    plt.grid(True)
    plt.show()
    
rk_plot_show(6,14)


```
